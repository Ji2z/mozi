```python
from PIL import Image
import os
import xml.etree.ElementTree as ET
```

```python
## 3개짜리 구분 여부
def include_three(file):
    name = file.split('_')
    if(name[2]=='s'):
        return False
    else:
        return True
```

```python
# 경로 저장
DIR_PATHS = './raw'

#IMAGE_PATHS = sorted(list(DIR_PATHS.glob("*.jpg")))
#XML_PATHS = sorted(list(DIR_PATHS.glob("*.xml")))

xml_list = []
img_list = []
file_list = os.listdir(DIR_PATHS)
isInclude = False # 3개짜리 사진도 포함할거면 True로 변경

for file in file_list:
    if '.xml' in file:
        if(isInclude == False):
            if(include_three(file) == False):
                xml_list.append(file)
        else:
            xml_list.append(file)
    elif '.jpg' in file:
        if(isInclude == False):
            if(include_three(file) == False):
                img_list.append(file)
        else:
            img_list.append(file)
```

```python
# 파일 경로 및 개수 확인
print(xml_list)
print(len(xml_list))
```

```python
## 이미지 사이즈 수정 : 2988 / 320 = 9.3375
num = 1
for img_file in img_list:
    img = Image.open(DIR_PATHS+"\\"+img_file)
    img_resize = img.resize((int(img.width / 9.3375), int(img.height / 9.3375)))
    img_resize.save("RESIZE/" + img_file)
    
    print("[" + str(num) + "]" + img_file + "[success]")
    num += 1
    
print("============ DONE = total : "+str(num)+" ============")
```
